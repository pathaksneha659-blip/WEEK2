Handling missing values:

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

Encoding:-


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df['Embarked']=le.fit_transform(df['Embarked'])
```

Standaradize:-


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])

df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15, None]
}
```

```python
rf = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Cross Validation Score:", grid_search.best_score_)
```

```python
import pandas as pd

best_rf = grid_search.best_estimator_
y_pred_test = best_rf.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred_test)

results = pd.DataFrame({
    "Best n_estimators": [grid_search.best_params_['n_estimators']],
    "Best max_depth": [grid_search.best_params_['max_depth']],
    "Cross Validation Score": [grid_search.best_score_],
    "Test Accuracy": [test_accuracy]
})

results
```

```python
import pandas as pd

results = pd.DataFrame({
    "Best n_estimators": [grid_search.best_params_['n_estimators']],
    "Best max_depth": [grid_search.best_params_['max_depth']],
    "Cross Validation Score": [grid_search.best_score_],
    "Test Accuracy": [test_accuracy]
})

results
```

GridSearchCV selected the best values for n_estimators and max_depth, but the model's accuracy remained almost the same. This indicates that the default Random Forest model was already performing well on the Titanic dataset.
