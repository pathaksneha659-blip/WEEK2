Import:-

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Handling missing values:

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

Encoding:-


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sex']=le.fit_transform(df['Sex'])
df['Embarked']=le.fit_transform(df['Embarked'])
```

Standaradize:-


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Age','Fare']] = scaler.fit_transform(df[['Age','Fare']])

df.head()
```

Calculate Precission,Recall and F1 score

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
precision=precision_score(y_test,model_DTC)
recall=recall_score(y_test,model_DTC)
f1=f1_score(y_test,model_DTC)
```

```python
precision_rf=precision_score(y_test,model_RF)
recall_rf=recall_score(y_test,model_RF)
f1_rf=f1_score(y_test,model_RF)
print("precision of Random Forest:",precision_rf)
print("recall of Random Forest:",recall_rf)
print("f1 of Random Forest:",f1_rf)
```

Now visualising confusion matrix using seaborn heatmap

```python
import seaborn as sns
```

```python
cm_rf = confusion_matrix(y_test, model_RF)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted Negative', 'Predicted Positive'],
            yticklabels=['Actual Negative', 'Actual Positive'])
plt.title('Confusion Matrix for Random Forest Classifier')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

creating  a summary table comparing accuracy,precission recall f1 score across all  3 models

```python
summary = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'Random Forest'],
    'Accuracy': [
        accuracy_score(y_test, model_LR),
        accuracy_score(y_test, model_DTC),
        accuracy_score(y_test, model_RF)
    ],
    'Precision': [
        precision_score(y_test, model_LR),
        precision_score(y_test, model_DTC),
        precision_score(y_test, model_RF)
    ],
    'Recall': [
        recall_score(y_test, model_LR),
        recall_score(y_test, model_DTC),
        recall_score(y_test, model_RF)
    ],
    'F1 Score': [
        f1_score(y_test, model_LR),
        f1_score(y_test, model_DTC),
        f1_score(y_test, model_RF)
    ]
})

summary
```

Best Performance Model and Worst Performed Model:

After evaluating all three models, Logistic Regression performed the best overall. It achieved the highest Accuracy (81.01%), Precision (78.57%), and F1-score (76.39%), showing that it was able to make more reliable predictions than the other models. It also maintained a good balance between identifying passengers who survived and those who did not.

On the other hand, the Decision Tree was the weakest-performing model. Although its Recall was similar to Logistic Regression, it had the lowest Accuracy, Precision, and F1-score. This means it made more incorrect predictions compared to the other two models.

The Random Forest performed better than the Decision Tree but was slightly behind Logistic Regression. While Random Forest generally helps reduce overfitting by combining multiple decision trees, in this case it achieved a slightly lower Recall and F1-score than Logistic Regression.

Reason for Difference in performance:

The performance differences are due to each model's learning approach and ability to generalize the data. As the model have their own condition performing.

Logistic Regression performed better because it generalized well on the Titanic dataset.

Decision Tree had lower performance as it is more prone to overfitting.

Random Forest reduced overfitting by combining multiple trees, but it still performed slightly below Logistic Regression on this dataset.
